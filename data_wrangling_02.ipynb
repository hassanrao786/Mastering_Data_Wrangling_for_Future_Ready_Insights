# 01-what is series?


####  A Pandas Series is like a column in a table.

#### It is a one-dimensional array holding data of any type

```python
import pandas as pd
```

```python
a=[1,2,3,4,5,63,1,233]
b=pd.Series(a)
print(b)
```

# 02 what is Data frame?


##### Data sets are usually multi-dimensional tables, called DataFrames.

##### Series is like a column, a DataFrame is the whole table.

##### A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

```python
a=pd.DataFrame({"age":[12,34,56,78,32,23],                                # making a data frame
                "height":[2.4,4.5,3.5,6.0,3.4,2.4]})
a
```

# 03-HOw  to rename a column?

```python
a.rename(columns={"age":"umar","height":"kad"})
```
